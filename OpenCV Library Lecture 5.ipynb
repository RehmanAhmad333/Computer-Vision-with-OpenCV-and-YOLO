## Where did YOLO come from?

Before YOLO:

Object detection was slow.

Old methods:

* Sliding window

Then in 2016:

Idea:

That is why name is:

## What problem does YOLO solve?

Old detection:

* Very slow
* Multiple passes
* High computation

YOLO:

* Single neural network
* Divides image into grid
* Predicts boxes + class at once
* Works in real-time

Problem solved:

‚ö° Speed + Accuracy together

## üîπ Step 2: What is Annotation?

Annotation = Drawing a box + giving label.

Example:

Image contains:

* Person wearing helmet
* Person without helmet

We draw:

Box ‚Üí ‚Äúhelmet‚Äù<br>
Box ‚Üí ‚Äúno_helmet‚Äù

This creates supervision.

YOLO does not see "helmet".

It sees:

` 
class_id x_center y_center width height
` 

Example:

` 
0 0.50 0.45 0.30 0.40
` 

Meaning:
```
Class 0
Box centered at (0.50, 0.45)
Width 30% of image
Height 40% of image
```
All values normalized (0 to 1).

## üîπ Step 3: How YOLO Actually Learns (Concept Level)

Inside training:

1. Input image goes into neural network
2. Network predicts:
   * Bounding boxes
   * Confidence score
   * Class probabilities
3. Compare with actual annotation
4. Calculate error (Loss Function)
5. Update weights
6. Repeat for many epochs`(Loops)`

After many repetitions:

Model becomes accurate.

## üîπ Step 4: YOLO Internal Working (Intermediate Level)

YOLO divides image into grid cells.

Each grid cell predicts:
* Box coordinates
* Object confidence
* Class probability


**1. Box coordinates**
Ye batata hai ke object **image me exactly kahan hai** ‚Äî yani bounding box ki position aur size.

Usually 4 values hoti hain:

* **x, y** ‚Üí box ka center
* **width, height** ‚Üí box ka size

Matlab: *object kis jagah par hai.*

---

**2. Object confidence**
Ye ek **score** hota hai jo batata hai:

* kya grid cell me koi object hai ya nahi
* prediction kitni confident hai

Confidence zyada ho ‚Üí model ko zyada yaqeen hai object present hai.

---

**3. Class probability**
Ye batata hai ke object **kaunsi category ka hai**.

Example:

* car
* person
* dog
* bicycle

Model har class ke liye probability deta hai.

---

Simple summary:

| Prediction        | Matlab             |
| ----------------- | ------------------ |
| Box coordinates   | Object kahan hai   |
| Object confidence | Object hai ya nahi |
| Class probability | Object kya hai     |



Final prediction:<br>
Bounding box + label + confidence score

#### Final prediction

Training ke baad model jo output deta hai usme 3 cheezein hoti hain:

**Bounding box + Label + Confidence score**

Matlab:

* **Bounding box** ‚Üí object kahan hai image me
* **Label** ‚Üí object kya hai (car, person, dog, etc.)
* **Confidence score** ‚Üí model ko kitna yaqeen hai

Example:

```
Car detected
Confidence: 0.92
Location: bounding box around car
```

Mathematically:<br>
Loss = Box loss + Confidence loss + Classification loss


### Loss function (training ke time)

Deep learning model ko train karne ke liye **loss function** use hota hai.
Loss basically batata hai model **kitni galti kar raha hai**.

Aap ne jo likha hai wo bilkul correct structure hai:

```
Loss = Box loss + Confidence loss + Classification loss
```

Ab inko samjhte hain.

---

### 1. Box loss

Ye measure karta hai:

> predicted bounding box actual box se kitna different hai

Agar box sahi jagah par nahi hai ‚Üí loss zyada hoga.

---

### 2. Confidence loss

Ye check karta hai:

> model ne object presence sahi predict ki ya nahi

Example:

* object hai lekin confidence low ‚Üí loss
* object nahi hai lekin confidence high ‚Üí loss

---

### 3. Classification loss

Ye measure karta hai:

> object ki class sahi predict hui ya nahi

Example:

* dog ko cat predict kar diya ‚Üí loss increase



### Deep learning optimization

Training ke time model repeatedly:

1. prediction karta hai
2. loss calculate hota hai
3. **weights update hote hain (backpropagation)**

Goal:

```
Loss ‚Üí minimum
Accuracy ‚Üí maximum
```

Isi process ko **deep learning optimization** kehte hain.

## Step 3: Create dataset.yaml file

# üîπ DATASET VALIDATION SCRIPT

Students must check before training. 



````python
import os

train_images = os.listdir("datasets/dataset/images/train")
train_labels = os.listdir("datasets/dataset/labels/train")

print("Train Images:", len(train_images))
print("Train Labels:", len(train_labels))
````

Counts must match.

## Advanced Check

Check missing label files:

````python
missing = []

for img in train_images:
    label_name = img.replace(".jpg", ".txt")
    if label_name not in train_labels:
        missing.append(img)

print("Missing labels:", missing)
````

## Step 4: Load Pretrained YOLO Model

````python
model = YOLO("yolov8n.pt")
````

Explanation:

* yolov8n = nano version (fast, small)
* Pretrained on COCO dataset

## Step 5: Start Training

````python
model.train(
    data = "C:/Users/User/OpenCV/datasets/dataset/dataset.yaml",
    epochs = 10,
    imgsz=640
)
````

Explanation line by line:

* data ‚Üí config file
* epochs ‚Üí number of full dataset passes
* imgsz ‚Üí image resize dimension

After training:

Model saved in:

```
runs/detect/train/weights/best.pt
```
